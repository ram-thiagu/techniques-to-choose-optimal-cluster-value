```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
#plt.style.available
```

```python
#reading dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
df1=df.copy()
```

```python
df.drop(['Response', 'id'], axis=1, inplace =True)
```

```python
df['Gender'] = df['Gender'].replace(('Male','Female'),(1, 0))
df['Vehicle_Age'] = df['Vehicle_Age'].replace(('< 1 Year','1-2 Year','> 2 Years'),(0,1,2))
df['Vehicle_Damage'] = df['Vehicle_Damage'].replace(('Yes','No'),(1, 0))
```

```python
from sklearn.cluster import KMeans
```

```python
# Weight of sum of squared distance
WSS = []
# For possible 'k' values 2-10
K = range(1, 10)
for n in K:
    algorithm = (KMeans(n_clusters = n) )
    algorithm.fit(df)
    WSS.append(algorithm.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, WSS, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
db = []
K = range(2, 10)
for n in K:
    algorithm = (KMeans(n_clusters = n) )
    algorithm.fit(df)
    labels = algorithm.labels_
    db.append(davies_bouldin_score(df,labels))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, db, 'bx-')
plt.xlabel('k')
plt.ylabel('DB index')
plt.title('The DB index Method showing the optimal k')
plt.show()
```

```python
df2=df1[df1['Response']==0].sample(10000)
df3=df1[df1['Response']==1].sample(10000)
```

```python
data = df2.append(df3)
```

```python
data.drop(['id', 'Response'], axis=1, inplace=True)
```

```python
data['Gender'] = data['Gender'].replace(('Male','Female'),(1, 0))
data['Vehicle_Age'] = data['Vehicle_Age'].replace(('< 1 Year','1-2 Year','> 2 Years'),(0,1,2))
data['Vehicle_Damage'] = data['Vehicle_Damage'].replace(('Yes','No'),(1, 0))
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil = []
# Number of clusters in range 2-10
K = range(2, 10)
for n in K:
    algorithm = (KMeans(n_clusters = n) )
    algorithm.fit(data)
    labels = algorithm.labels_
    sil.append(silhouette_score(data, labels))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, sil, '^b-')
plt.xlabel('k')
plt.ylabel('silhouette_score')
plt.title('The silhouette method showing the optimal k')
plt.show()
```
